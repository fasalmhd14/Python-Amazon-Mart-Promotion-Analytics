```python
# Generate a report that displays each campaign along with the total revenue generated before and after the total campaign? 
#The report includes three key fields:
# campaign_name
# total_revenue_before_campaign
# total_revenue_after_campaign
#(Display the value in millions)
# This report should help in evaluating the financial impact of our promotional campaigns.
```

```python
df_events.head()
```

```python
#Revenue before and after promo
df_events = df_events
df_events['revenue_bf'] = df_events['base_price'] * df_events['quantity_sold(before_promo)']
df_events['revenue_af'] = df_events['base_price'] * df_events['quantity_sold(after_promo)']
df_events.head()
```

```python
#aggregate revenue per campaigns
# create the group by data frame object for campaign category
df_events_gp = df_events.groupby(['campaign_id'])
df_events_gp
```

```python
df_report_3 =df_events_gp[['revenue_bf', 'revenue_af']].sum().reset_index()
df_report_3
```

```python
#Converting the values into millions
df_report_3['revenue_bf'] = df_report_3['revenue_bf']/1000000
df_report_3['revenue_af']  = df_report_3['revenue_af']/1000000
df_report_3
```

```python
df_report_3 = pd.merge(df_campaigns, df_report_3, on = 'campaign_id')
df_report_3 = df_report_3.iloc[:,[1,4,5]]
df_report_3
```

```python
# 4. Generate a report that calculates the Incremental sold quantity(ISU%) for each category during the diwali campaign. 
# Additionally, provide rankings for the categories, based on their ISU%.
# Required Fields:
# Category, 
# ISU %,
# rank 

# Note : ISU%(Incremental sold quantity percentage) is calculated as the increase/decrease in 
# quantity sold after promo compared to quantity sold before promo.
```

```python
df_events.head()
```

```python
df_events['ISU%'] = (((df_events['quantity_sold(after_promo)'] - df_events['quantity_sold(before_promo)']) / df_events['quantity_sold(before_promo)']) * 100).round(2)
df_events.head()
```

```python
#merging the product table with event table
df_prod_event = pd.merge(df_products, df_events, on='product_code')
df_prod_event.head()
```

```python
#SELECTING RECORDS FROM ONLT DIWALI CAMPAIGHN
df_prod_event = df_prod_event[df_prod_event['campaign_id'] == "CAMP_DIW_01"]
df_prod_event
```

```python
# creating a dataframe group by object on 'category' column 
```

```python
df_category_gp = df_prod_event.groupby(['category'])
df_category_gp
```

```python
#calulating the average icu% for the group 'category'
df_report_4 = df_category_gp['ISU%'].mean().reset_index().round(2)
df_report_4
```

```python
df_report_4['rankings'] = df_report_4.rank(ascending = False, numeric_only = True)
df_report_4.sort_values(['rankings'], inplace = True, ignore_index = True)
df_report_4
```

```python
# 5. Create a report featuring the top 5 products , ranked by incremental revenue percentage(IR%), across all campaigns
# Required Fields:
# Product name,
# Category
# IR%
# This analysis helps identify the most successful products in terms of incremental revenue across our campaigns, assisting in product optimization.
```

```python
df_events.head()
```

```python
#Deriving the column for IR%
df_events['IR%'] = (((df_events['revenue_af'] - df_events['revenue_bf']) / df_events['revenue_bf']) * 100).round(2)
df_events.head()
```

```python
#merging the product table with new event table
df_prod_event_2 = pd.merge(df_products, df_events, on = 'product_code')
df_prod_event_2.head()
```

```python
# creating a dataframe group by object on 'product_code' column 
```

```python
df_product_gp = df_prod_event_2.groupby(['product_name'])
df_product_gp
```

```python
#calulating the average IR% for the group 'product_id'
df_report_5 = df_product_gp['IR%'].mean().reset_index().round(2)
df_report_5
```

```python
df_report_5['rankings'] = df_report_5.rank(ascending = False, numeric_only = True)
df_report_5.sort_values(['rankings'], inplace = True, ignore_index = True)
df_report_5 = df_report_5[0:5]
df_report_5
```
